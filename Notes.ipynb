## Data type

```python
print("Hello"[4])
```

If you want to add numbers together, should not use strings, but integer - whole numbers; then it will add them together.
-Large number- can have _ in it for read easily, but will be skipped by the interpreter

```python
print("123" + "456")
print(123 + 456)
print(f"Large number: {123_456}")
print(123.456)
print(True, False)
```
